```python
#@title 4: This is the Hack button. Press it.
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
import fade
import textwrap
import json
import requests
import urllib.parse
import urllib.request
import argparse
import sys
import os
import json
import csv
import shutil
from os import path
from shutil import make_archive
from directory_structure import Tree
from alive_progress import alive_bar
import time
import openai
from dotenv import load_dotenv

load_dotenv()  
apiToken = os.environ.get("OPENAI_TOKEN")
headers = {
                    "Accept": "application/json; charset=utf-8",
                    "Authorization": "Token" + str(apiToken)
                }
hack_dir = 'hackGPT/'
hack_input = 'input/'
hack_output = 'output/'
targets = hack_input+"hax"
investigation = hack_output+"hax"
search = open(targets ,"r")
query = search.read()

query_parse = json.dumps(query.split("\n"), sort_keys=True, indent=4)

print("\033[36mBypassing Moderation endpoint...")
print("\033[36mSuccess!")
print("\033[36mSending below query to OpenAI:\n")
print(fade.brazil(query))
seperator = "─ ──+──── ──  ─   ────────────────────────────────────────────────────────   ─  ── ────+── ─"
faded_seperator = fade.purplepink(seperator)
print(faded_seperator)
openai_targets = []

if 'OPENAI_TOKEN' in os.environ:
    pass
else:
    os.environ['OPENAI_TOKEN'] = input('Enter API Key: ').replace(" ","")

f_jsonpath = investigation+'/results'
token = os.environ.get("OPENAI_TOKEN")

if os.path.exists(investigation):
  pass
else:
  os.mkdir(investigation)

if os.path.exists(investigation+'/results/'):
  pass
else:
  os.mkdir(investigation+'/results/')
with open(targets, 'r') as targets:
    for line in targets:
        openai_targets.append(line)

    for search in openai_targets:
        response = openai.Completion.create(
        model="text-davinci-003",
        prompt="\n\n1."+search,
        temperature=0,
        max_tokens=3000,
        top_p=1,
        frequency_penalty=0,
        presence_penalty=0,
        stop=["\"\"\""]
        )
        response = response['choices'][0]['text']
        if os.path.exists(investigation+"/results/hackGPT.txt"):
          with open(investigation+"/results/hackGPT.txt", "a") as f:
            f.write("\n----------------------------------------------------\n")
            f.write("Query:")
            f.write("\n")
            f.write(search)
            f.write("\nResponse:")
            f.write(response) 
        else:
          with open(investigation+"/results/hackGPT.txt", "a") as f:
            f.write("\n----------------------------------------------------\n")
            f.write("Query:")
            f.write("\n")
            f.write(search.strip('\n'))
            f.write("\nResponse:")
            f.write(response.strip('\n')) 
        fadedresponse = fade.fire(str(response))
    print(' '*39+"🆁🅴🆂🆄🅻🆃🆂\n" )
    print(fadedresponse)
    print(faded_seperator)
    
path = Tree(investigation+'/results/', absolute=True)
print(path)
with open(investigation+"/results/hackGPT.txt", 'r') as logfile:
    print("Current Log:")
    head = [next(logfile) for x in range(20)]
    for x in head:
        print(x)
    

```
